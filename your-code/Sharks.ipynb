# 3.15 Year

```python
df.columns
```

```python

# Fill NaN values in the 'Year' column with 0.0
df['Year'].fillna(0.0, inplace=True)

# Convert the 'Year' column to numeric (integer) type
df['Year'] = pd.to_numeric(df['Year'], errors='coerce').astype('Int64')

# Define the regular expression pattern for a 4-digit year
formato_bueno = r'(\d{4})'

# Apply a function to update 'Year' if it's 0 and a valid year can be found in the 'Date' column
df['Year'] = df.apply(lambda row: int(re.search(formato_bueno, row['Date']).group(1)) if pd.isna(row['Year']) and re.search(formato_bueno, row['Date']) else row['Year'], axis=1)
```

```python
df_als=df_als[~(df_als['Sex']=='unknown')]
```

```python
# Agrupar por 'Country' y 'Sex' y contar los casos 'Y' en 'Fatal (Y/N)'
grouped = df_als.groupby(['Country', 'Sex'])['Fatal (Y/N)'].count()
# Crear el gráfico de barras
colores_pastel = ['#FFCCCC', '#A9DFBF']
grouped.unstack().plot(kind='bar', stacked=True, color=colores_pastel)
plt.title('Número de ataques por Género en United States y Australia')
plt.xlabel('País y Género')
plt.ylabel('Número de ataques')
plt.legend(title='Sexo')
plt.show()
```

```python
grouped = grouped.unstack()
grouped_percentage = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Crear el gráfico de barras con porcentajes
colors = ['lightpink','lightblue']
grouped_percentage.plot(kind='barh', stacked=True, colormap='Pastel1', color=colors)

plt.title('Porcentaje de Muertes por Género en United States y Australia')
plt.xlabel('Porcentaje de Muertes')
plt.ylabel('País y Género')
plt.legend(title='Sexo')
plt.show()
```
