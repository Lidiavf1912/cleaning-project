# 3.7 Fatal (Y/N)

```python
df[df['Injury'].isna()]
```

```python
df['Injury'] = df['Injury'].astype(str)
df['Fatal (Y/N)'] = df['Fatal (Y/N)'].astype(str)
```

```python
for i, row in df.iterrows():
    if "FATAL" in row['Injury'] or 'Y' in row['Fatal (Y/N)']:
        df.at[i, 'Fatal (Y/N)'] = 'Y'
    else:
        if 'N' in row['Fatal (Y/N)']:
            df.at[i, 'Fatal (Y/N)'] = 'N'
        else:
            df.at[i, 'Fatal (Y/N)'] = 'UNKNOWN'
```

```python
df['Fatal (Y/N)'].unique()
```

```python
for i, row in df.iterrows():
    if 'Y' in row['Fatal (Y/N)']:
        df.at[i, 'Injury'] = 'FATAL'
    else:
        pass
```
